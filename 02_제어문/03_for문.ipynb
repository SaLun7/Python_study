```python
# for문의 기본 구조
# for 변수 in 리스트(또는 튜플, 문자열):
#     수행할 문장1
#     수행할 문장2
```

```python
test_list = ["one", "two", "three"]

for tl in test_list:
    print(tl)
```

```python
a = [(1, 2), (3, 4), (5, 6)]

for k, j in a:
    print(k, j)
```

```python
marks = [90, 25, 67, 45, 80]
number = 0

for mark in marks:
    number = number + 1
    if mark >= 60:
        print(f"{number} 학생은 합격입니다.")
    else:
        print(f"{number} 학생은 불합격입니다.")
```

```python
# for문 continue

marks = [90, 25, 67, 45, 80]
number = 0

for mark in marks:
    number += 1
    if mark < 60:
        continue

    print(f"{number}학생은 합격입니다.")
```

```python
# for 문과 함께 자주 사용하는 range 함수

a = range(10)
print(list(a))
```

```python
b = range(1, 10 + 1, 2)
print(list(b))
```

```python
c = range(10, 0, -1)
print(list(c))
```

```python
# for와 range를 이용한 구구단

for i in range(2, 4):
    for j in range(1, 9 + 1):
        print(f"{i} X {j} = {i * j}")
```

```python
# 리스트 컴프리핸션 사용하기

a = [1, 2, 3, 4]
result = []

for num in a:
    result.append(num * 3)

print(result)
```

```python
# for문과 break문

for i in range(10): #range(0, 10)
    if i == 5:
        break
    print(i)
```

```python
# zip함수로 여러 리스트 순회하기

names = ['홍길동', '김철수', '이영희']
scores = [85, 92, 78]

for name, score in zip(names, scores):
    print(f"{name}: {score}점")
```
