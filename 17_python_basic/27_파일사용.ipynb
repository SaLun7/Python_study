```python
# 파일에 문자열 쓰기, 읽기

# 파일객체 = open(파일이름, 파일모드)

# 파일객체.write("문자열")

# 파일객체.close()

file = open("./data/hello.txt", "w")
file.write("hello world!")
file.close()
```

```python
## 파일에서 문자열 읽기

f = open("./data/hello.txt", "r")
s = f.read()
print(s)
f.close()
```

```python
## 자동으로 파일 객체 닫기

# with open(파일이름, 파일모드) as 파일객체:
#     코드

with open("data/hello.txt", "r") as file:
    s = file.read()
    print(s)
```

```python
# 문자열 여러 줄을 파일에 쓰기, 읽기

with open("./data/hello.txt", "w") as file:
    for i in range(3):
        file.write(f"hello world {i}\n")
```

```python
lines = [
    "안녕하세요.\n",
    "파이썬\n",
    "코딩 도장입니다.\n"
]

with open("./data/hello.txt", "w") as f:
    f.writelines(lines)
```

```python
## 파일의 내용을 한 줄씩 리스트로 가져오기

with open("./data/hello.txt", "r") as file:
    lines = file.readlines()
    print(lines)
```

```python
## 파일의 내용을 한 줄씩 읽기

with open("./data/hello.txt", "r") as f:
    line = None
    while line != "":
        line = f.readline()
        print(line.strip("\n"))
```

```python
## for 반복문으로 파일의 내용을 줄 단위로 읽기

with open("./data/hello.txt", "r") as f:
    for line in f:
        print(line.strip("\n"))
```

```python
# 파이썬 객체를 파일에 저장하기, 가져오기

import pickle

name = "james"
age = 17
address = "서울시 서초구 반포동"
scores = {
    "Korean" : 90,
    "English" : 95,
    "Mathematics" : 85,
    "Science" : 82
}

with open("./data/james.p", "wb") as file:
    pickle.dump(name, file)
    pickle.dump(age, file)
    pickle.dump(address, file)
    pickle.dump(scores, file)
```

```python
import pickle

with open("./data/james.p", "rb") as file:
    name = pickle.load(file)
    age = pickle.load(file)
    address = pickle.load(file)
    scores = pickle.load(file)

    print(name)
    print(age)
    print(address)
    print(scores)
```

```python
with open("./data/words.txt.", "r") as file:
    count = 0
    words = file.readlines()
    for word in words:
        if len(word.strip("\n")) <= 10:
            count += 1

    print(count)
```
