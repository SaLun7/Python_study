```python
# 문자열 조작하기
## 문자열 바꾸기

print("Hello, World".replace("World", "Python"))
```

```python
## 문자바꾸기

table = str.maketrans("aeiou", "12345")
print("apple".translate(table))
```

```python
## 문자열 분리하기
text = "apple pear grape pineapple ornage"
print(text.split())
```

```python
text = "apple, pear, grape, pineapple, ornage"
print(text.split(","))
```

```python
text = "apple, pear, grape, pineapple, ornage"
print(text.split(", "))
```

```python
# 구분자 문자열과 문자열 리스트 연결하기

lis = ['apple', 'pear', 'grape', 'pineapple', 'ornage']

print(" ".join(lis))
print("-".join(lis))
```

```python
## 소문자를 대문자로, 대문자를 소문자로
print("Python".upper())
print("Python".lower())
```

```python
## 공백삭제하기
print("  Python  ".strip())
print("  Python  ".rstrip())
print("  Python  ".lstrip())
```

```python
## 특정문자 삭제하기
print(", Python.".strip(",."))
```

```python
print(", !Python.!".strip(",.!"))
print(", !Python.!".strip(",.! "))
print("*, !Python.!".strip(",.! "))
```

```python
## 문자열을 왼쪽 정렬하기
print("Python".ljust(10))
```

```python
print("Python".rjust(10))
```

```python
## 문자열을 가운데 정렬하기
print("Python".center(10))
```

```python
## 매서드 체이닝
print("Python".rjust(10).upper())
print("Python".rjust(10).upper().strip())
```

```python
## 문자열 왼쪽에 0 채우기
print("35".zfill(4))
print("3.5".zfill(6))
print("hello".zfill(10))
```

```python
## 문자열 위치 찾기

print("apple pineapple".find("pl"))
```

```python
print("apple pineapple".find("xy"))
```

```python
print("apple pineapple".rfind("pl"))
```

```python
# 문자열 서식 지정자와 포메팅 사용하기

## 서식 지정자로 문자열 넣기
# "%s" % "문자열"

print("I am %s" % "james")
```

```python
name = "maria"
print("I am %s" % name)
```

```python
## 서식 지정자로 숫자넣기
# "%d" % 숫자

print("I am %d years old" % 20)
```

```python
## 서식 지정자로 소수점 표현하기
# "%f" % 숫자

print("%f" % 2.3)
```

```python
# "%.자리수f" % 숫자

print("%.2f" % 2.3)
print("%.3f" % 2.3)
print("%10.3f" % 2.3)
```

```python
## 서식지정자로 문자열 정렬하기

print("%10s" % "Python")
print("%-10s" % "Python")
```

```python
## 서식지정자로 문자열 안에 값 여러개 넣기

print("Today is %d %s" % (3, "April"))
```

```python
## format 메서드 사용하기

print("hello {0}".format("world"))
print("hello {0} {1} {2}".format("Python", "Script", 3.6))
```

```python
print("{0} {0} {1} {1}".format("Python", "Script"))
```

```python
print("Hello {language} {version}".format(language = "Python", version = 3.6))
```

```python
## format 메서드로 문자열 정렬하기
print("{0:<10}".format("Python"))
print("{0:>10}".format("Python"))
print("{:>10}".format("Python"))
```

```python
## 숫자 개수 맞추기
print("%03d" % 1)

print("{0:03d}".format(35))
```

```python
print("%08.2f" % 3.6)
```

```python
print("{0:08.2f}".format(3.6))
```
