```python
# 변수 여러개를 한 번에 만들기
x, y, z = 10, 20, 30

print(x)
print(y)
print(z)
```

```python
# x, y, z = 10, 20 (x)
```

```python
x = y = z = 10
print(x)
print(y)
print(z)
```

```python
x, y = 10, 20
x, y = y, x
print(x)
print(y)
```

```python
# 변수로 계산하기
a = 10
b = 20
print(10 + 20)
```

```python
# 산술 연산 후 할당 연산자 사용하기
a = 10
a + 20
print(a)
```

```python
b = 10
b += 20
print(b)
```

```python
# 입력값을 정수로 반환하기

a = int(input("첫 번째 숫자를 입력하세요. : "))
b = int(input("두 번째 숫자를 입력하세요. : "))

print(a + b)
```

```python
# 입력 값을 변수 두 개에 저장하기

a, b = input("문자열 두 개를 입력하세요. : ").split()
```

```python
print(a)
print(b)
```

```python
print("hello world".split())
```

```python
a, b = ['hello', 'world']
print(a)
print(b)
```

```python
a, b = int(input("숫자 두 개를 입력하세요. : ")).split() #(x)
print(a)
print(b)
```

```python
a, b = input("숫자 두 개를 입력하세요. : ").split() #(o)
a = int(a)
b = int(b)

print(a + b)
```

```python
print(list(map(int, ["10", "20"])))
```

```python
a, b = map(int, input("숫자 두 개를 입력하세요. : ").split())

print(a)
print(b)
print(a + b)
```

```python
# 입력받은 값을 콤마를 기준으로 분리하기

a, b = map(int, input("숫자 두 개를 입력하세요. : ").split(","))
print(a + b)
```

```python
print("10,20".split())
print("10,2,0".split(","))
```

```python
a, b = map(int, input("숫자 두 개를 입력하세요. : ").split())
print(a + b)
```

```python
a, b, c = map(int, input().split())
print(a + b + c)
```

```python
a, b, c = map(str, input().split())

print(a)
print(b)
print(c)
```

```python
a, b, c, d = list(map(int, input().split()))
f = int((a + b + c + d) / 4)
print(f)
```
