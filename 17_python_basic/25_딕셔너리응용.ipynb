```python
# 딕셔너리 조작하기
## setdefault, update

a = {
    "a" : 10,
    "b" : 20,
    "c" : 30,
    "d" : 40
}
```

```python
a.setdefault("e")
print(a)
```

```python
a.setdefault("f", 100)
print(a)
```

```python
## 딕셔너리의 키의 값 수정

a = {
    "a" : 10,
    "b" : 20,
    "c" : 30,
    "d" : 40
}
```

```python
a.update(a=90)
print(a)
```

```python
a.update(e=90)
print(a)
```

```python
a.update(a=900, f=60)
print(a)
```

```python
y = {
    1 : "one",
    2 : "two"
}

y.update(
    {1 : "ONE", 2 : "TWO"}
)

print(y)
```

```python
y.update(
    [
        [2, "two"],
        [4, "four"]
    ]
)

print(y)
```

```python
y.update(
    zip(
        [1, 2],
        ["one", "two"]
    )
)

print(y)
```

```python
## 딕셔너리 표현식 사용하기

# {키:값 for 키, 값 in 딕셔너리}
# dict({키:값 for 키, 값 in 딕셔너리})


keys = ["a", "b", "c", "d"]
imsi = {
    "a" : None,
    "b" : None,
    "c" : None,
    "d" : None
}
# {for key, value in dict.fromkeys(keys).items()}
x = {key:value for key, value in imsi.items()}

print(x)
```

```python
b = {value : 0 for value in {
    "a" : 10,
    "b" : 20,
    "c" : 30,
    "d" : 40
}.values()
}

print(b)
```

```python
## 딕셔너리 표현식에서 if 조건문 사용하기

x = {
    "a" : 10,
    "b" : 20,
    "c" : 30,
    "d" : 40
}

for key, value in x.items():
    if value == 20:
        del x[key]

print(x)
```

```python
x = {
    "a" : 10,
    "b" : 20,
    "c" : 30,
    "d" : 40
}

x = {key: value for key, value in x.items() if value != 20}
```
