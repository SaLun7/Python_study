```python
# import 모듈 가져오기

# import 모듈
# import 모튤1, 모듈2

# 모듈.변수
# 모듈.함수()
# 모듈.클래스()

import math

math.pi
math.sqrt(4.0)
```

```python
## import as로 모듈 이름 지정하기

# import 모듈 as 이름

import math as m

print(m.sqrt(4.0))
print(m.pi)
```

```python
# from import 로 모듈의 일부만 가져오기

# import math

from math import pi, sqrt

print(pi)
print(sqrt(4))
```

```python
# from 모듈 import *

from math import *
```

```python
# from import로 모듈의 일부를 가져온 뒤 이름 지정하기

from math import sqrt as s, pi as p

print(s(4.0))
print(p)
```
