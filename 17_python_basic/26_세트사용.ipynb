```python
## 집합연산 사용하기

a = {1, 2, 3, 4}
b = {3, 4, 5, 6}


print(a | b)
print(a.union(b))
print(set.union(a, b))
```

```python
print(a & b)
print(a.intersection(b))
print(set.intersection(a, b))
```

```python
print(a - b)
print(a.difference(b))
print(set.difference(a, b))
```

```python
### 대칭차집합

print(a ^ b)
print(set.symmetric_difference(a, b))
```

```python
# 반복문으로 세트의 요소를 모두 출력하기

a = {1, 2, 3, 4}

for i in a:
    print(i)
```

```python
# 세트 표현식 사용하기

# {식 for 변수 in 반복 가능한 객체}
# set(식 for 변수 in 반복 가능한 객체)

a = {i for i in "apple"}
print(a)
```

```python
## 세트 표현식에 if조건문 사용하기

# {식 for in 세트 if 조건식}
# set({식 for in 세트 if 조건식})

a = {i for i in "pineapple" if i not in "apl"}

print(a)
```

```python
a = {i for i in range(1, 101) if i % 3 == 0}
b = {j for j in range(1, 101) if j % 5 == 0}

print(a & b)
print(a.intersection(b))
```

```python
num1, num2= map(int, input().split())

a = {i for i in range(1, num1+1) if num1 % i == 0}
b = {j for j in range(1, num2+1) if num2 % j == 0}


divisor = a & b

result = 0
if type(divisor) == set:
    result = sum(divisor)

print(result)
```
