```python
# 인덱스와 요소를 함께 출력하기 -enumerate()

a = [38, 21, 53, 62, 19]

for value in enumerate(a):
    print(value)
```

```python
a = [38, 21, 53, 62, 19]

for index, value in enumerate(a):
    print(index, value)
```

```python
a = [38, 21, 53, 62, 19]

for index, value in enumerate(a):
    print(f"{index + 1}번째 값은 {value}입니다.")
```

```python
# while문으로 요소 출력하기

a = [38, 21, 53, 62, 19]

i = 0
while i < len(a):
    print(a[i])
    i += 1
```

```python
# 리스트의 가장 작은 수, 가장 큰 수, 함께 구하기
# min(), max(), sum()

a = [38, 21, 53, 62, 19]
smallest = a[0]

for i in a:
    if i < smallest:
        smallest = i

print(smallest)
```

```python
a = [38, 21, 53, 62, 19]
largest = a[0]

for i in a:
    if i > largest:
        largest = i

print(largest)
```

```python
a = [38, 21, 53, 62, 19]

x = 0

for i in a:
    x = x + i

print(x)
```

```python
# 리스트 표현식 사용학
# [식 for 변수 in 리스트]
# list(식 for 변수 in 리스트)

a = [i for i in range(10)]
print(a)
```

```python
b = list(i for i in range(10))
print(b)
```

```python
c = [i + 5 for i in range(10)]
print(c)
```

```python
# 리스트 표현식에서 if조건문 사용하기

# [식(변수) for 변수 in 리스트 if 조건식]
# list([식(변수) for 변수 in 리스트 if 조건식]

a = [i for i in range(10) if i % 2 == 0]
print(a)
```

```python
a = []
for i in range(10):
    if i % 2 == 0:
        a.append(i)
print(a)
```

```python
b = [i + 5 for i in range(10) if i % 2 == 1]
print(b)
```

```python
# for 반복문과 if조건문을 여러 번 사용하기
a = [i * j for i in range(2, 10) for j in range(1, 10)]
print(a)

b = []
for i in range(2, 10):
    for j in range(1, 10):
        b.append(i * j)

print(b)
```

```python
#  리스트에 map 사용하기

# list(map(함수, 리스트))
# tuple(map(함수, 튜플))

a = [1.2, 2.5, 3.7, 4.6]

for i in range(len(a)):
    a[i] = int(a[i])

print(a)
```

```python
a = [1.2, 2.5, 3.7, 4.6]

a = list(map(int, a))
print(a)
```
