```python
# 사각형으로 별 출력하기

for i in range(5):
    for j in range(5):
        print("*", end="")
    print()
```

```python
for _ in range(3):
    for _ in range(7):
        print("*", end="")
    print()
```

```python
# 계단식으로 별 출력하기

for i in range(5):
    for j in range(5):
        if j <= i:
            print("*", end="")
    print()
```

```python
# 대각선으로 별 출력하기

for i in range(5):
    for j in range(5):
        if j == i:
            print("*",end="")
        else:
            print(" ",end="")
    print()
```

```python
h = int(input())

for i in range(h):
    for j in reversed(range(h)):
        if j > i:
            print(" ", end="")
        else:
            print("*", end="")

    for j in range(h):
        if j < i:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```

```python
h = int(input())

for i in range(h):
    for _ in range(h - i - 1):
        print(" ", end="")

    for _ in range(2 * i + 1):
        print("*", end="")
    
    print()
```

```python
h = int(input())

for i in range(h):
    print(" " * (h - i -1) + "*" * (2 * i + 1))
```

```python
height = 5

for i in range(1, height + 1):
    space = height - i
    star = 2 * i - 1
    print(" " * space + "*" * star)
```

```python
n = int(input())

for i in range(n):

    for _ in range(n - i - 1):
         print(" ", end="")

    for _ in range(2 * i + 1):
         print("*",end="")

    print()

for j in range(n - 1):
    for _ in range(j + 1):
         print(" ", end="")

    for _ in range(2*(n - j - 1) - 1):
         print("*", end="")
    
    print()
    
```
