```python
# 함수에서 값을 여러 개 반환하기

def add_sub(a, b):
    return a + b, a - b

x, y = add_sub(10, 20)
print(x)
print(y)
```

```python
# 함수의 호출 과정 알아보기

def mul(a, b):
    c = a * b
    return c

def add(a, b):
    c = a + b
    print(c)
    d = mul(a, b)
    print(d)

x = 10
y = 20
print(add(x, y))
```

```python
def get_quotitent_remainder(a ,b):
    return a // b, a % b
    

x = 10
y = 3

quotient, remainder = get_quotitent_remainder(x ,y)
print("몫: {0}, 나머지: {1}".format(quotient, remainder))
```

```python
x, y = map(int, input().split())

def calc(a, b):
    return a + b, a - b, a * b, a / b

a, s, m, d = calc(x, y)
print("덧셈: {0}, 뺄셈: {1}, 곱셈: {2}, 나눗셈: {3}".format(a, s, m, d))
```
