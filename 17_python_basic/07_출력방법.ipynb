```python
# 값을 여러개 출력하기
print(1, 2, 3)
print("Hello", "Python")
```

```python
#  sep로 값 사이에 문자 넣기
print(1, 2, 3, sep = "")
print(4, 5, 6, sep = ',')
print(1920, 1000, sep = " X ")
```

```python
print(1, 2, 3, sep = "\n")
print("1\n2\n3")
```

```python
# end 사용하기
print(1)
print(2)
print(3)
```

```python
print(1, end = "")
print(2, end = "")
print(3, end = "")
```

```python
print(1, end = " ")
print(2, end = " ")
print(3, end = " ")
```

```python
year = 2000
month = 10
day = 27
hour = 11
minute = 43
second = 59

print(year, month, day, sep = "/", end= " ")
print(hour, minute, second, sep = ":")
```

```python
year, month, day, hour, minute, second = input().split()
print(year, month, day, sep = "-", end = "T")
print(hour, minute, second, sep = ":")
```
