```python
# 재귀호출 사용하기

def hello():
    print("Hello World!")
    hello()

hello()
```

```python
## 재귀호출에 종료 조건 만들기

def hello(count):
    if count == 0:
        return
    
    print("Hello World!", count)

    count -= 1
    hello(count)

hello(5)
```

```python
# 재귀호출로 팩토리얼 구하기

def factorial(n):
    if n == 1:
        return 1
    
    return n * factorial(n - 1)
            # 5 * factorial(4)
            #     4 * factorial(3)
            #         3 * factorial(2)
            #             2 * factorial(1)
            #                 1

print(factorial(5))
```

```python
def is_palindrome(word):

    if len(word) < 2:
        return True

    if word[0] != word[-1]:
        return False
    
    return is_palindrome(word[1:-1])

print(is_palindrome("hello"))
print(is_palindrome("level"))
```

```python
def fib(n):
    if n <= 1:
        return n
    else:
        return fib(n - 1) + fib(n - 2)

n = int(input())
print(fib(n))
```
