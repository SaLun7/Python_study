```python
# 문자열은 어떻게 만들고 사용할까?

## 1. 큰 따옴표로 양쪽 둘러싸기
a = "Hello worild"

## 2. 작은 따옴표로 양쪽 둘러싸기
b = 'Python is fun'

## 3. 큰 따옴표 3개로 둘러싸기
c = """Life is too short, You need Python"""

## 4. 작은 따옴표 3개로 둘러싸기
d = '''Life is too short, You need Python'''

print(a)
print(b)
print(c)
print(d)
print(type(a))
print(type(b))
print(type(c))
print(type(d))
```

```python
# 문자열안에 작은 따옴표 포함하기
a = "Python's favorite food is perl"
b = """Python's favorite food is perl"""
c = 'Python\'s favorite food is perl'
print(c)
```

```python
# 문자열 안에 큰 따옴표 포함하기

a = '"Python is very easy" he says.'
b = "\"Python is very easy\" he says."

print(b)
```

```python
mulitline = """Life is too short, 
you need Python."""
mulitline2 = "Life is too short, \nyou need Python."

print(mulitline)
print(mulitline2)
```

```python
# 문자열 포매팅
## 1. 숫자 바로 대입

print("I eat %d apples" % 3)

## 2. 문자열 바로 대입

print("I eat %s apples" % "five")
```

```python
## 3. 숫자값을 나타내는 변수로 대입

number = 3
print("I eat %d apples" % number)
```

```python
## 2개 이상의 값 넣기

number = 10
day = "three"
print("I eat %d apples so I was sick for %s days" % (number, day))
```

```python
# format 함수를 사용한 포매팅

print("I eat {0} apples".format(3))
print("I eat {0} apples".format("five"))
```

```python
# format 함수를 이용해서 2개이상의 값 넣기

number = 10
day = "three"
print("I eat {0} apples so I was sick for {1} days".format(number, day))
```

```python
# f 문자열 포매팅

name = "홍길동"
age = 10
print(f"나의 이름은 {name}입니다. 나이는 {age}입니다.")
```

```python
# 문자열 관련 함수들

a = "hobby"
print(type(a))

## 문자 개수 세기 - count
print(a.count("b"))
```

```python
## 위치 알려주기 - find

a = "Python is the best choice"
print(a.find("b"))
print(a.find("k"))
```

```python
# 위치 알려주기2 - index

a = "Life is too short"
print(a.index("t"))
print(a.index("k"))
```

```python
# 문자열 삽입 - join

print(",".join("abcd"))
```

```python
# 소문자를 대문자로 바꾸기 - upper
a = "hi"
print(a.upper())
```

```python
# 대문자를 소문자로 바꾸기 - lower

a = "HI"
print(a.lower())
```

```python
# 공백지우기

a = "   hi   "
print(a)
print(a.strip())
```

```python
# 공백지우기 왼쪽 / 오른쪽

a = "   hi   "
print(a.lstrip())
print(a.rstrip())
```

```python
# 문자열 바꾸기 - replace

a = "Life is too short"
print(a.replace("Life", "your leg"))
```

```python
# 문자열 나누기 - split

a = "Life is too short"
print(a.split())
```
